```python
# loading libraries
import pandas as pd
import numpy as np
import nltk
import os
import nltk.corpus# sample text for performing tokenization
from nltk.tokenize import word_tokenize# Passing the string text into word tokenize for breaking the sentences
import re
from langdetect import detect
```

```python
# pre-prepare
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# manipulating data
file_name = '1st-gen.txt'
f = open(file_name,"r")
lines = f.read()
f.close()
def d(line):
    try:
        return detect(line)
    except:
        return "-"

lines = [ l for l in lines.split('\n') if d(l) == 'en' ]
lines = ' '.join(lines)
lines = re.sub("(http|ftp|https):\/\/([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:\/~+#-]*[\w@?^=%&\/~+#-])"," ",lines.lower())
lines = re.sub("[^a-zA-Z]"," ",lines.lower())
lines = re.sub("[ ]+"," ",lines.lower())
lines = re.sub("smart contract","smartcontract",lines.lower())
lines = re.sub("smart contracts","smartcontract",lines.lower())
lines = re.findall('\w{2,}', lines)
lines = ' '.join([x for x in lines])
w = open('man-'+file_name, "w")
w.write(lines)
w.close()
```

```python
# tokenization
from nltk import word_tokenize
from nltk.corpus import stopwords
file = 'man-1st-gen.txt'
f = open(file, "r")
text = f.read()
f.close()
a = set(stopwords.words('english'))
token1 = word_tokenize(text.lower())
token = [x for x in token1 if x not in a]
token

```

```python
# Importing Porterstemmer from nltk library
# Checking for the specific words
from nltk.stem import PorterStemmer
pst = PorterStemmer()
token = [ pst.stem(x) for x in token ]
```

```python
# getting frequency by word
from nltk.probability import FreqDist
ignore_words = ["dao","https", "rt", "de", "la", "https", "el", "los", "en", "un", "que", "via", "amp", "se", "las", "great", "good", "back", "con"]
man_token = [x for x in token if x not in ignore_words]
fdist = FreqDist(man_token)
fdist1 = fdist.most_common(60)
fdist1
```
